Import libraries and read data

```python
import pandas as pd
import numpy as np
import math
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("/content/drive/MyDrive/01Exercise1.csv")
print(df.head())
df.shape
```

```python
#Data Pre processing
#Make a copy for analysis if needed at a later time
df_prep=df.copy()

#Find missing row count
df_prep.isnull().sum(axis=0)
```

```python
#dop na as count is not significant
df_prep=df_prep.dropna()
df_prep.shape
```

```python
df_prep.isnull().sum(axis=0)
```

```python
#gender is not a reasonable column to consider,can be dropped
df_prep=df_prep.drop(['gender'],axis=1)
```

```python
#create dummy variables for married,ch(credit history),status. 
df_prep.dtypes
```

```python
df_prep=pd.get_dummies(df_prep,drop_first=True)
```

```python
#normalize loanamt and income
#normalization is ideally done on train data as the 1st step after split, and transformed on test data. hence this is an exception(for practice only) 
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df_prep['income']=scaler.fit_transform(df_prep[['income']])
df_prep['loanamt']=scaler.fit_transform(df_prep[['loanamt']])
df_prep.head()
```

```python
#Build Logistic Regression
X=df_prep.drop('status_Y',axis=1)
y=df_prep['status_Y']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,stratify=y,random_state=1234)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
#Build confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
lr.score(X_test,y_test)
```
